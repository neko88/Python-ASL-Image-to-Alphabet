```python
# Setup for environment

import matplotlib.pyplot as plt
import numpy as np
import PIL
import os
import tensorflow as tf
from tensorflow import keras
from keras.utils import load_img, img_to_array
from keras.utils import Sequence
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '1'
```

# Custom Generator
To trim the number of images of each class
Returns the image and label

```python
class CustomImageGenerator(Sequence):
    def __init__(self, directory, classes, images_per_class, target_size=(150, 150), batch_size=32):
        self.directory = directory
        self.classes = classes
        self.images_per_class = images_per_class
        self.target_size = target_size
        self.batch_size = batch_size
        self.filepaths = []
        self.labels = []
        self._initialize()

    def _initialize(self):
        # for each class name in specified classes...
        for class_name in self.classes:
            # get path of the class name [i]
            class_dir = os.path.join(self.directory, class_name)
            # get the img names in the class name [i] path
            image_files = os.listdir(class_dir)
            np.random.shuffle(image_files)
            # take only specified num images
            selected_images = image_files[:self.images_per_class]
            
            for img_name in selected_images:
                # get path for x, images
                self.filepaths.append(os.path.join(class_dir, img_name))
                # get labels for y, alphabet
                self.labels.append(self.classes.index(class_name))
        
        # convert to np. arrays for x and y arrays
        self.filepaths = np.array(self.filepaths)
        self.labels = np.array(self.labels)

    def __len__(self):
        return int(np.ceil(len(self.filepaths) / float(self.batch_size)))

    def __getitem__(self, idx):
        batch_x = self.filepaths[idx * self.batch_size:(idx + 1) * self.batch_size]
        batch_y = self.labels[idx * self.batch_size:(idx + 1) * self.batch_size]

        # Iterate the image path, load, process, and append.
        images = []
        for img_path in batch_x:
            img = load_img(img_path, target_size=self.target_size)
            img_array = img_to_array(img)
            images.append(img_array)

        # Scale the images and store them in the array
        images = np.array(images).astype('float32') / 255.0
        batch_y = np.array(batch_y)

        return images, batch_y
```

```python
import cv2

def extract_num_images(directory, classes, image_per_class, target_size):
    filepaths = []
    labels = []
# for each class name in specified classes...
    for class_name in classes:
        # get path of the class name [i]
        class_dir = os.path.join(directory, class_name)
        # get the img names in the class name [i] path
        image_files = os.listdir(class_dir)
        np.random.shuffle(image_files)
        # take only specified num images
        selected_images = image_files[:images_per_class]

        for img_name in selected_images:
            # get path for x, images
            filepaths.append(os.path.join(class_dir, img_name))
            # get labels for y, alphabet
            labels.append(classes.index(class_name))

    # convert to np. arrays for x and y arrays
    filepaths = np.array(filepaths)
    labels = np.array(labels)
    
        # Iterate the image path, load, process, and append.
    images = []
    for img_path in filepaths:
        img = cv2.imread(img_path)
        img = cv2.resize(img, target_size)
        img = img_to_array(img)
        img = np.array(img)
        img = np.reshape(img, (1,150,150,2))
        img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        images.append(img_grey)

    # Scale the images and store them in the array
    images = np.array(images).astype('float32') / 255.0
        
    return images, labels
```

# Use the CustomImageGenerator to extract num. images per class
Each index of the object contain the specified batch_size images

```python
train_dir = 'img/asl_alphabet_train'
class_names = os.listdir(train_dir)
class_names.pop(0)
images_per_class = 100
num_classes = len(class_names)

train_imgs = extract_num_images(train_dir, class_names, images_per_class, (150,150))
```

```python
# Show some images from a batch
batch_images, batch_labels = train_imgs

# Display the first 5 images in the batch
for i in range(5):
    print(batch_images[i].shape)
    plt.imshow(batch_images[i])
    plt.title(f"Label: {batch_labels[i]}")
    plt.show()
```

```python
from keras.preprocessing.image import ImageDataGenerator
test_dir = 'img/asl_alphabet_test'

test_datagen = ImageDataGenerator(rescale=1.0/255.0)  # Normalizing the images if needed

# Load images using flow_from_directory
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),  # Adjust to match your model's input size
    batch_size=32,  # Adjust as needed
    class_mode=None,  # Since it's test data, you typically don't have labels
    shuffle=False,  # Keep the order of images as in the directory
    classes=None  # Since all images are in one folder, no need to specify subfolders
)

# Check the number of images loaded
print(f"Number of test images: {test_generator.samples}")
```

# Create and Compile the Model

```python
def create_model(input_shape, num_classes):
    model = Sequential()
    
    # Convolutional layers
    model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Conv2D(64, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
    
    model.add(Conv2D(128, (3, 3), activation='relu'))
    model.add(MaxPooling2D((2, 2)))
   
    # Fully connected layers
    model.add(Flatten())
    model.add(Dense(512, activation='relu'))
    model.add(Dropout(0.5))
    
    model.add(Dense(num_classes, activation='softmax'))
    
    # Compile the model
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    return model
```

```python
model = create_model(input_shape=(150,150), num_classes=num_classes)
```

# Train the Model

```python
epochs = 1
history = model.fit(train_imgs[0],train_imgs[1], epochs=epochs, steps_per_epoch=len(train_imgs))
```

```python
print(history.history)
```

```python
acc = history.history['accuracy']
loss = history.history['loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Accuracy')
plt.legend(loc='lower right')
plt.title('Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Loss')
plt.legend(loc='upper right')
plt.title('Loss')
plt.show()
```
